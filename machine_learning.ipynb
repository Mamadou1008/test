```python
from sklearn import preprocessing
```

```python
transform = preprocessing.LabelEncoder()
```

```python
base_bdd = pd.read_excel("BDD_prets.xls", sheet_name = "pret") # Import d'un fichier excel + onglet spécifique
```

```python
base_bdd.info()
```

```python
colonnes_quali = base_bdd.select_dtypes(include = "object").columns
```

```python
colonnes_quali
```

```python
base_ess = base_bdd.copy()
```

```python
base_ess #BDD d’apprentissage
```

```python
for colonne in colonnes_quali:
    base_ess[colonne]=transform.fit_transform(base_ess[colonne])
```

```python
base_ess
```

```python
base_ess_colonn=base_ess.dropna(axis=1) #supprimer les colonnes qui contiennent des données manquantes
```

```python
arbre = tree.DecisionTreeClassifier()
X= base_ess_colonn.loc[:,base_ess_colonn.columns!="Succès"]
Y= base_ess_colonn["Succès"]
#arbre=arbre.fit(X,Y)
#tree.plot_tree(arbre)
```

```python
#Y = Y.reshape(Y.shape[0], 1)
```

```python
#Découpage de la base de données BDD
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.66)
```

### SVM

```python
#Classe svm de Scikit-Learn
#Sous-classes 
#svm.SVC pour une classification et 
#svm.SVR pour une régression
```

```python
from sklearn.svm import SVC # classification
from sklearn.svm import SVR # régression
```

```python
#classifier = SVC(kernel='linear’) 
classifier=SVC(kernel='poly', degree=5)
# ou (kernel='rbf’) pour un noyau gaussien
classifier.fit(X_train, Y_train)
Y_pred = classifier.predict(X_test) 
# idem pour SVR
```

### Evaluation du modéle SVM

```python
#Classe metrics de sklearn
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test,Y_pred)) 
print(classification_report(Y_test,Y_pred))
```
